## Shortest Way to Form String

From any string, we can form a subsequence of that string by deleting some number of characters (possibly no deletions).

Given two strings source and target, return the minimum number of subsequences of source such that their concatenation equals target. If the task is impossible, return -1.

Example 1:
```
Input: source = "abc", target = "abcbc"
Output: 2
```

Explanation: The target "abcbc" can be formed by "abc" and "bc", which are subsequences of source "abc".
Example 2:
```
Input: source = "abc", target = "acdbc"
Output: -1
```
Explanation: The target string cannot be constructed from the subsequences of source string due to the character "d" in target string.

Example 3:
```
Input: source = "xyz", target = "xzyxz"
Output: 3
```
Explanation: The target string can be constructed as follows "xz" + "y" + "xz".
 

Constraints:

Both the source and target strings consist of only lowercase English letters from "a"-"z".
The lengths of source and target string are between 1 and 1000.

````python
def shortestWay(self, source: str, target: str) -> int:
    if len(set(list(target)) - set(list(source))):
        return - 1
    target += "!"
    i, j = 0, 0
    ans = 0
    while j < len(target)-1:
        if target[j] == source[i]:
            j += 1
            i += 1
        while i < len(source) and target[j] != source[i]:
            i += 1
        if i == len(source):
            ans += 1
            i = 0
    return ans
````

````python
import pysnooper
@pysnooper.snoop()
def shortestWay_MN(source, target):
    ans = 0
    i = 0
    while i < len(target):
        tmp = i
        for j in range(len(source)):
            if i < len(target) and source[j] == target[i]:
                i += 1
        if i == tmp:
            return -1
        ans += 1
    return ans
````

````python
shortestWay_MN("abcd", "aaad")
````

````python
import bisect
@pysnooper.snoop()
def shortest_way_to_form_string(source, target):
    letter_to_occurences = [[] for _ in range(26)]
    for i in range(len(source)):
        letter_to_occurences[ord(source[i]) - ord('a')].append(i)

    count = 0
    prev = -1
    for i in range(len(target)):
        occurences = letter_to_occurences[ord(target[i]) - ord('a')]
        if not occurences:
            return - 1
        idx = bisect.bisect_left(occurences, prev)
        if idx >= len(occurences):
            count += 1
            next_source_idx = occurences[0] + 1
        else:
            next_source_idx = occurences[idx] + 1
        prev = next_source_idx
    return count
````

````python
shortest_way_to_form_string("abcd", "aaad")
````
